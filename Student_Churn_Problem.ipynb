```python
# Joining both data file w.r.t user_id
can_data = pd.merge(eng, can, how='right', on='user_id')
```

```python
# Lets see how many rows and columns data have
can_data.shape
```

```python
# Deriving KPI by calculating total content view per user
uq1= []
uq1 = can_data.user_id.unique()
result = []
for x in uq1:
    temp = can_data.loc[can_data['user_id']==x, ['content_views']]
    con_views = temp.content_views.sum() 
    result.append((x, con_views))

views_wise = pd.DataFrame(result,columns=['user_id','content_views'])
can_data1 = pd.merge(can_data, views_wise, how='left', on='user_id')
```

```python
# Converting date columns in standard format
can_data1['cancel_date'] = pd.to_datetime(can_data1['cancel_date'],format='%Y-%m-%d')
can_data1['order_created_date'] = pd.to_datetime(can_data1['order_created_date'],format='%Y-%m-%d')
can_data1['most_recent_billing_date'] = pd.to_datetime(can_data1['most_recent_billing_date'],format='%Y-%m-%d')
```

```python
# Deriving KPI - total subsciption time held by per user in months
can_data1['subs_tenure'] = can_data1['cancel_date'] - can_data1['order_created_date']
can_data1['subs_tenure'] = round(can_data1['subs_tenure']/np.timedelta64(1,'M'))
```

```python
can_data1
```

```python
# Dropping variables not required
can_data1 = can_data1.drop(['date', 'order_line_id_x','content_views_x','order_line_id_y'], axis=1)
```

```python
can_data1.head(10)
```

```python
# Eliminating Duplicates from Data
can_data1 = can_data1.drop_duplicates()
```

```python
# How data size is-
can_data1.shape
```

```python
can_data1.head(10)
```

```python
# Assuming null values under engagement_platform as both(Mobile&Web)
can_data1['engagement_platform'] = can_data1['engagement_platform'].fillna('Both')
```

```python
#6. In Data Table, creating a new variable called â€œTenure_Categoryâ€ using subs_tenure as follows. -  
#- Long Active - > 24 months  
#- Active â€“ Between 2 and 12  
#- On Trial â€“ less than or equal to 0 

bins = [0, 2, 12, 100]
seg_names = ['On Trial','Active', 'Long Active']
can_data1['Tenure_Category'] = pd.cut(can_data1['subs_tenure'], bins, labels = seg_names, include_lowest=True)
can_data1
```

```python
# Grouping columns to develop some insights
count = can_data1.groupby(['engagement_platform', 'cancel_type']).size() 
```
